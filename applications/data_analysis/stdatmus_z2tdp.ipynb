# Atmosphere Properties above Sea-Level

## Overview

The properties of the atmosphere change with the height and this notebook will cover:
- International Standard Atmosphere
- `ambiance` Python package

## International Standard Atmosphere

The International Standard Atmosphere represents a static model for how the atmosphere changes over a range of altitiudes above sea-level. The atmosphere is split into different named layers, where properties like temperature, pressure, and density change at predictable rates.

## [`ambiance`](https://ambiance.readthedocs.io/en/latest/) Python Package

There are different models to represent how the atmosphere changes with height. The older 1976 NASA _U.S Standard Atmosphere_ and the more recent 1993 International Civil Aviation Organization (ICAO) _Manual of the ICAO Standard Atmosphere_ which extends the efficacy of the 1976 model to 80 km. The `ambiance` package makes use of the 1993 model, but will [match the 1976 model in most parts](https://ambiance.readthedocs.io/en/latest/theory/references.html). `ambiance` supports altitudes from -5004 m to 81020 m measured at sea-level.

The `ambiance` package will return the atmosphere properties:
- Density (`density`)
- Pressure (`pressure`)
- Temperature (`temperature_in_celsius`)
- Collision frequency (`collision_frequency`)
- Dynamic viscosity (`dynamic_viscosity`)
- Geometric height above [mean sea-level](https://en.wikipedia.org/wiki/Sea_level) (`h`)
- Geopotential height (`H`)
- Gravitational acceleration (`grav_accel`)
- Kinematic viscosity (`kinematic_viscosity`)
- Layer names (`layer_name`)
- Mean free path (`mean_free_path`)
- Mean particle speed (`mean_particle_speed`)
- Number density (`number_density`)
- Pressure scale height (`pressure_scale_height`)
- Specific weight (`specific_weight`)
- Speed of sound (`speed_of_sound`)
- Temperature (`temperature`)
- Thermal conductivity (`thermal_conductivity`)

### Troposphere: 0 - 11,000 meters

The troposphere represents the lowest layer of the Earth's atmospehre. This layer starts at ground level and represents where humans live and most weather occurs.

```python
from ambiance import Atmosphere

height = 1655  # meters above sea-level (city of Boulder)
atmos = Atmosphere(height)

print(f"Layer Name     : {atmos.layer_name[0]}")
print(f"Density        : {atmos.density[0]} kg/m^3")
print(f"Pressure       : {atmos.pressure[0]/100} hPa (mB)")
print(f"Temperature    : {atmos.temperature_in_celsius[0]} C")
print(f"Speed of sound : {atmos.speed_of_sound[0]} m/s")
```

### Stratosphere: 20,000 - 47,000 m

The stratosphere is where the ozone layer is found and a region where the temperature increases instead of decreasing.

```python
from ambiance import Atmosphere

height = 21054  # meters above sea-level
atmos = Atmosphere(height)

print(f"Layer Name     : {atmos.layer_name[0]}")
print(f"Density        : {atmos.density[0]} kg/m^3")
print(f"Pressure       : {atmos.pressure[0]/100} hPa (mB)")
print(f"Temperature    : {atmos.temperature_in_celsius[0]} C")
print(f"Speed of sound : {atmos.speed_of_sound[0]} m/s")
```

### Stratopause: 47,000 - 51,000 m

The stratopause is a boundary level between the stratosphere and the mesophere, where the peak temperature of the stratosphere occurs.

```python
from ambiance import Atmosphere

height = 51000  # meters above sea-level
atmos = Atmosphere(height)

print(f"Layer Name     : {atmos.layer_name[0]}")
print(f"Density        : {atmos.density[0]} kg/m^3")
print(f"Pressure       : {atmos.pressure[0]/100} hPa (mB)")
print(f"Temperature    : {atmos.temperature_in_celsius[0]} C")
print(f"Speed of sound : {atmos.speed_of_sound[0]} m/s")
```

### Mesosphere: 51,000 - 84,852 m

The mesosphere is an upper layer of the atmosphere where meteors typically burn up and where the coldest temperature in the Earth's atmosphere is found.

```python
from ambiance import Atmosphere

height = 61000  # meters
atmos = Atmosphere(height)

print(f"Layer Name     : {atmos.layer_name[0]}")
print(f"Density        : {atmos.density[0]} kg/m^3")
print(f"Pressure       : {atmos.pressure[0]/100} hPa (mB)")
print(f"Temperature    : {atmos.temperature_in_celsius[0]} C")
print(f"Speed of sound : {atmos.speed_of_sound[0]} m/s")
```

### Collect Changes in Atmosphere Properties with Altitude

Properties like density, pressure, temperature, and the speed of sound all change with the height above sea-level in the atmosphere.

```python
from ambiance import Atmosphere
import numpy as np

height_range = np.arange(-5004, 81020, 100)
density = []
pressure = []
temperature = []
speed_of_sound = []

for height in height_range:
    atmos = Atmosphere(height)
    density.append(atmos.density[0])
    pressure.append(atmos.pressure[0] / 100)
    temperature.append(atmos.temperature_in_celsius[0])
    speed_of_sound.append(atmos.speed_of_sound[0])
```

### Plot Change in Density as Altitude Increases

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 10))
plt.title("Change in Density as Altitude Increases")
plt.xlabel("Height (m)")
plt.ylabel("Density (kg/m^3)")
plt.plot(height_range, density)
plt.show()
```

### Plot Change in Pressure as Altitude Increases

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 10))
plt.title("Change in Pressure as Altitude Increases")
plt.xlabel("Height (m)")
plt.ylabel("Pressure (hPa mB)")
plt.plot(height_range, pressure)
plt.show()
```

### Plot Change in Temperature as Altitude Increases

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 10))
plt.title("Change in Temperature as Altitude Increases")
plt.xlabel("Height (m)")
plt.ylabel("Temperature (C)")
plt.plot(height_range, temperature)
plt.show()
```

### Plot Change in Speed of Sound as Altitude Increases

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 10))
plt.title("Change in Speed of Sound as Altitude Increases")
plt.xlabel("Height (m)")
plt.ylabel("Speed of Sound (m/s)")
plt.plot(height_range, speed_of_sound)
plt.show()
```

## Calculating Anomalies

xarray's [`groupby()`](https://docs.xarray.dev/en/stable/generated/xarray.DataArray.groupby.html) function can be used to split the data on various timescales.

```python
from pythia_datasets import DATASETS
import xarray as xr
import matplotlib.pyplot as plt

filepath = DATASETS.fetch("CESM2_sst_data.nc")
ds = xr.open_dataset(filepath)
tos_monthly = ds.tos.groupby("time.month")
tos_clim = tos_monthly.mean(dim="time")
tos_anom = tos_monthly - tos_clim
tos_anom
```

```python
# Plot the anomalies at the first time
tos_anom.isel(time=0).plot()
```

## Removing Annual Cycle

Also known as deseasonalization

```python
from pythia_datasets import DATASETS
import xarray as xr

filepath = DATASETS.fetch("CESM2_sst_data.nc")
ds = xr.open_dataset(filepath)
tos_monthly = ds.tos.groupby("time.month")
tos_clim = tos_monthly.mean(dim="time")
tos_anom = tos_monthly - tos_clim
tos_anom
```

```python
# Plot the global mean TOS anomaly
tos_anom.mean(dim=["lat", "lon"]).plot()
plt.ylabel("Global mean tos anomaly")
plt.xlabel("Time (months)")
```

## Calculating Long Term Means

```python
from pythia_datasets import DATASETS
import xarray as xr
import matplotlib.pyplot as plt

filepath = DATASETS.fetch("CESM2_sst_data.nc")
ds = xr.open_dataset(filepath)
tos_monthly = ds.tos.groupby("time.month")
tos_clim = tos_monthly.mean(dim="time")
tos_clim
```

```python
# Plot the long term mean for a sample location
tos_clim.sel(lon=310, lat=50, method="nearest").plot()
```

## Calculating Seasonal Means

```python
from pythia_datasets import DATASETS
import xarray as xr
import matplotlib.pyplot as plt

filepath = DATASETS.fetch("CESM2_sst_data.nc")
ds = xr.open_dataset(filepath)
tos_seasonal = ds.tos.groupby("time.season")
tos_clim = tos_seasonal.mean(dim="time")
tos_clim
```

```python
# Plot the JJA seasonal mean
tos_clim.sel(season="JJA").plot()
```

## Finding The Standard Deviation of Monthly Means

```python
from pythia_datasets import DATASETS
import xarray as xr
import matplotlib.pyplot as plt

# Load the dataset
filepath = DATASETS.fetch("CESM2_sst_data.nc")
ds = xr.open_dataset(filepath)
stdMon = ds.tos.groupby("time.month").std()
stdMon
```

```python
stdMon.sel(month=1).plot()
```

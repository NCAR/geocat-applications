### addfile
NCL's [`addfile`](https://www.ncl.ucar.edu/Document/Functions/Built-in/area_poly_sphere.shtml) opens a data file from supported file formats

#### Grab and Go

````python
# netCDF and HDF5 and HDF-EOS5
import geocat.datafiles as gcd
import xarray as xr

netCDF_datafile = xr.open_dataset(gcd.get("netcdf_files/1994_256_FSD.nc"))
netCDF_datafile
````

Hierarchical Data Format (HDF) and NASA Earth Observing System (EOS) data products

By default, `xarray` supports HDF5 files, but by specifying the engine, HDF files can be opened as a missing dependency

````python
# HDF and HDF-EOS
import geocat.datafiles as gcd
import xarray as xr

hdf_datafile = xr.open_dataset(gcd.get("hdf_files/avhrr.hdf"), engine="netcdf4")
hdf_datafile
````

GRIB requires the additional `cfgrib` engine option, which can be installed alongside as:
```
conda install -c conda-forge cfgrib eccodes
```

````python
# GRIB: Gridded Binary
import geocat.datafiles as gcd
import xarray as xr

grib_datafile = xr.open_dataset(gcd.get("grib_files/POT.grb"), engine="cfgrib")
grib_datafile
````

````python
# Shapefile
import geocat.datafiles as gcd
import geopandas as gpd

from osgeo import gdal

gdal.SetConfigOption('SHAPE_RESTORE_SHX', 'YES')

shp_datafile = gpd.read_file(gcd.get("shape_files/states.shp"))
shp_datafile.head()
````

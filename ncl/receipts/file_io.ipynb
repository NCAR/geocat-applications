## Python Functionality

````python
import geocat.datafiles as gcd
import xarray as xr
from osgeo import gdal
import geopandas as gpd # shapefiles

# netCDF file format family: netCDF, HDF5, and HDF-EOS5
netCDF_datafile = xr.open_dataset(gcd.get("netcdf_files/1994_256_FSD.nc"))
assert isinstance(netCDF_datafile, xr.Dataset), "netCDF file failed to load into Xarray"
assert netCDF_datafile, "netCDF files failed to load"
assert not netCDF_datafile is None, "netCDF file is empty"

CDF_datafile = xr.open_dataset(gcd.get("netcdf_files/941110_P.cdf"))
assert isinstance(CDF_datafile, xr.Dataset), "CDF file failed to load into Xarray"
assert CDF_datafile, "CDF files failed to load"
assert not CDF_datafile is None, "CDF file is empty"

# Non-HDF5 file format family: HDF and HDF-EOS
HDF_datafile = xr.open_dataset(gcd.get("hdf_files/avhrr.hdf"), engine="netcdf4")
assert isinstance(HDF_datafile, xr.Dataset), "HDF file failed to load into Xarray"
assert HDF_datafile, "HDF files failed to load"
assert not HDF_datafile is None, "HDF file is empty"

HE2_datafile = xr.open_dataset(gcd.get("hdf_files/MOD04_L2.A2001066.0000.004.2003078090622.he2"), engine="netcdf4")
assert isinstance(HE2_datafile, xr.Dataset), "he2 file failed to load into Xarray"
assert HE2_datafile, "he2 files failed to load"
assert not HE2_datafile is None, "he2 file is empty"

# GRIB file format family
grib_datafile = xr.open_dataset(gcd.get("grib_files/POT.grb"), engine="cfgrib")
assert isinstance(grib_datafile, xr.Dataset), "GRIB file failed to load into Xarray"
assert grib_datafile, "GRIB files failed to load"
assert not grib_datafile is None, "GRIB file is empty"

grb2_datafile = xr.open_dataset(gcd.get("grib_files/200711210135_NCV_CEILING_A.grb2"), engine="cfgrib")
assert isinstance(grb2_datafile, xr.Dataset), "grb2 file failed to load into Xarray"
assert grb2_datafile, "grb2 files failed to load"
assert not grb2_datafile is None, "grb2 file is empty"

# Shapefile file format family
shp_datafile = gpd.read_file(gcd.get("shape_files/states.shp"))
assert isinstance(shp_datafile, gpd.GeoDataFrame), "shp file failed to load into GeoPandas"
assert not shp_datafile.empty, "shp is empty"

shx_datafile = gpd.read_file(gcd.get("shape_files/countyl010g.shx"))
assert isinstance(shx_datafile, gpd.GeoDataFrame), "shx file failed to load into GeoPandas"
assert not shx_datafile.empty, "shx is empty"
````

## Functions

### addfile
NCL's [`addfile`](https://www.ncl.ucar.edu/Document/Functions/Built-in/area_poly_sphere.shtml) opens a data file from supported file formats.

Supported file formats:
- netCDF
- HDF
- HDF5
- HDF-EOS
- HDF-EOS5
- CCM
- GRIB
- OGR

#### Grab and Go

````python
# netCDF
# file extensions: .nc, cdf, or .netcdf
import geocat.datafiles as gcd
import xarray as xr

netCDF_datafile = xr.open_dataset(gcd.get("netcdf_files/1994_256_FSD.nc"))
netCDF_datafile
````

By default, `xarray` supports HDF5 files, but by specifying the engine, HDF files can be opened as a missing dependency

````python
# HDF and HDF-EOS
# file extension: .hdf or .hd
# file extension: .hdfeos, .he2, or .he4
# Hierarchical Data Format (HDF) for all NASA Earth Observing System (EOS) data products
import geocat.datafiles as gcd
import xarray as xr

hdf_datafile = xr.open_dataset(gcd.get("hdf_files/avhrr.hdf"), engine="netcdf4")
hdf_datafile
````

````python
# HDF5 and HDF-EOS5
# file extension: .h5 or .he5
import geocat.datafiles as gcd
import xarray as xr

hdf_datafile = xr.open_dataset(
    gcd.get("hdf_files/3B-MO.MS.MRG.3IMERG.20140701-S000000-E235959.07.V03D.HDF5")
)
hdf_datafile
````

````python
# GRIB: Gridded Binary
# file extensions: .gr, .gr1, .grb, .grib, .grb1, .grib1, .gr2, .grb2, or .grib2
import geocat.datafiles as gcd
import xarray as xr

grib_datafile = xr.open_dataset(gcd.get("grib_files/POT.grb"), engine="cfgrib")
grib_datafile
````

## Functions

### addfile
NCL's [`addfile`](https://www.ncl.ucar.edu/Document/Functions/Built-in/area_poly_sphere.shtml) opens a data file from supported file formats

#### Grab and Go

- netCDF and HDF5 and HDF-EOS5

````python
import geocat.datafiles as gcd
import xarray as xr

netCDF_datafile = xr.open_dataset(gcd.get("netcdf_files/1994_256_FSD.nc"))
netCDF_datafile
````

#### Grab and Go

- HDF and HDF-EOS

By default, `xarray` supports HDF5 files, but can be updated to a specified engine to work with additional files formats

````python
import geocat.datafiles as gcd
import xarray as xr

hdf_datafile = xr.open_dataset(gcd.get("hdf_files/avhrr.hdf"), engine="netcdf4")
hdf_datafile
````

#### Grab and Go

- GRIB

GRIB requires the additional `cfgrib` engine, which can be installed alongside Xarray:
```
conda install -c conda-forge cfgrib eccodes
```

````python
import geocat.datafiles as gcd
import xarray as xr

grib_datafile = xr.open_dataset(gcd.get("grib_files/ST4.2002030112.06h.grb"), engine="cfgrib")
grib_datafile
````

#### Grab and Go

- Shapefiles

`gdal` sets `SHAPE_RESTORE_SHX` to `YES` to repair a missing `.shx` index file for the ESRI Shapefile

````python
# Shapefile
import geocat.datafiles as gcd
import geopandas as gpd

from osgeo import gdal

gdal.SetConfigOption('SHAPE_RESTORE_SHX', 'YES')

shp_datafile = gpd.read_file(gcd.get("shape_files/states.shp"))
shp_datafile.head()
````

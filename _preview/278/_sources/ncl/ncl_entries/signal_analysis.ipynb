### cfftf
NCL's [`cfftf`](https://www.ncl.ucar.edu/Document/Functions/Built-in/cfftf.shtml) calculates an unnormalized forward complex discrete fourier transform of a real periodic sequence

```python
import numpy as np

example_data = [1002, 1017, 1018, 1020, 1018, 1027,
                1028, 1030, 1012, 1012, 982, 1012,
                1001, 996, 995, 1011, 1027, 1025,
                1030, 1016, 996, 1006, 1002, 982]

cfftf = np.fft.fft(example_data)

print(cfftf.real)
```
